```python
def solve_n_queens(n):
    board = [[0 for _ in range(n)] for _ in range(n)]
    solutions = []

    def is_safe(row, col):
        for i in range(row):
            if board[i][col] == 1:
                return False
        i, j = row, col
        while i >= 0 and j >= 0:
            if board[i][j] == 1: return False
            i -= 1
            j -= 1   
        i, j = row, col
        while i >= 0 and j < n:
            if board[i][j] == 1: return False
            i -= 1
            j += 1 
        return True
    def backtrack(row):
        if row == n:
            solutions.append([" ".join(['Q' if x == 1 else '.' for x in r]) for r in board])
            return
        for col in range(n):
            if is_safe(row, col):
                board[row][col] = 1  
                backtrack(row + 1)   
                board[row][col] = 0  
    backtrack(0)
    return solutions
n = int(input("Enter the value of N: "))
results = solve_n_queens(n)

print(f"\n--- Printing all {len(results)} solutions for N={n} ---")
print("-" * 30)
for i, sol in enumerate(results, 1):
    print(f"Solution #{i}:")
    for row in sol:
        print(row)
    print("-" * 30)
print(f"\nTOTAL SOLUTIONS FOUND: {len(results)}")
```
